```python
from bs4 import BeautifulSoup
import requests
```

```python
page =requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi/all-rounder")
page
```

```python
page.content
```

```python
Soup=BeautifulSoup(page.content)
Soup
```

```python
Soup.prettify
```

```python
#player_name
n=Soup.find('div', class_="rankings-block__banner--name-large")#top position playe having seprate tag and class for playes column.
N=n.text.replace('\n','')
N
```

```python
Name=N.split('\n')
Name
```

```python
#team Name
t=Soup.find('div', class_="rankings-block__banner--nationality")#top position playe having seprate tag and class for playes column.
T=t.text.replace('\n','')
T
```

```python
TEAM=T.split()
TEAM
```

```python
#CAREER BEST RATING
R=Soup.find('span', class_="rankings-block__career-best-text")
cbr=R.text.replace('\n','')
cbr
```

```python
Carrer_best_rating=cbr.split('\n')
Carrer_best_rating
```

```python
#Rating
rate=Soup.find('td',class_="u-text-left")
Rat=rate.text.replace('\n','')
Rat
```

```python
RATING=Rat.split()
RATING
```

# 1st DataFrame
(for top most player)

```python
import pandas as pd
df_1=pd.DataFrame({})
df_1['PLAYER']=Name
df_1['TEAM']=TEAM
df_1['CAREER BEST RATING']=Carrer_best_rating
df_1['RATING']=RATING
df_1
```

# 2nd DataFrame

```python
#Player Name
nam=Soup.find_all('td', class_="table-body__cell rankings-table__name name")# here the tag and class are same so we applyed find_all and scraped all data at a single time.
nam
```

```python
PN=[]
for i in nam:
    PN.append(i.text.replace('\n',''))
PN
```

```python
#TEAM
t=Soup.find_all('td', class_="table-body__cell nationality-logo rankings-table__team")
t
```

```python
TEAM=[]
for i in t:
    TEAM.append(i.text.replace('\n',''))
TEAM
```

```python
#CAREER BEST RATING
bestR=Soup.find_all('td', class_="table-body__cell u-text-right u-hide-phablet")
bestR
```

```python
CBR=[]
for i in bestR:
    CBR.append(i.text.replace('\n',''))
CBR
```

```python
#rating
R=Soup.find_all('td', class_="table-body__cell rating")
R
```

```python
RATING=[]
for i in R:
    RATING.append(i.text.replace('\n',''))
RATING
```

```python
#creating the 2nd dta frame
df2=pd.DataFrame({})
df2['PLAYER']=PN
df2['TEAM']=TEAM
df2['CAREER BEST RATING']=CBR
df2['RATING']=RATING
df2
```

```python
#combine Data Frame 1 and 2
Womens_ODI_All_Rounder_Rankings=pd.concat([df_1,df2],ignore_index=True)
Womens_ODI_All_Rounder_Rankings
```

# top 10 womens ODI all-rounder
