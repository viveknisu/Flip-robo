```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
soup.prettify
```

```python
#name
name=soup.find_all('span',class_="u-hide-phablet") #team columns is having sametag and calss name for all so we apply find_all
name
```

```python
TEAM=[]
for i in name:
    TEAM.append(i.text.replace('\n',''))
TEAM
```

```python
FirstTEAM=TEAM[0:1]
FirstTEAM
```

```python
#Matches
m=soup.find('td', class_="rankings-block__banner--matches")# we wapply find function because tag and calss name for MATCH column is in this case id diffrent form others
M=m.text.replace('\n','')
M
```

```python
match=M.split()
match
```

```python
#POINTS
P=soup.find('td', class_="rankings-block__banner--points")
point=P.text.replace('\n','')
point
```

```python
Point=point.split()
Point
```

```python
#RATING
r=soup.find('td', class_="rankings-block__banner--rating u-text-right")
R=r.text.replace('\n','')
R
```

```python
Rate=R.split()
Rate
```

# 1st DataFrame

```python
#we wil create ist data frame for the top position team which is having same tage and calss in TEAMS columns but in rest others columns the tag and classes for are diffrent from others teams .
import pandas as pd
df_1=pd.DataFrame({})
df_1['TEAM']=FirstTEAM
df_1['MATCHES']=match
df_1['POINTS']=Point
df_1['RATING']=Rate
df_1
```

# TEAM

```python
k='Australia' # we will seprate the top postion team from list
TEAMS=[i for i in TEAM if i!=k ]
TEAMS
```

```python
T=TEAMS[0:9]
T
```

# MATCHES & POINTS

```python
m=soup.find_all('td', class_="table-body__cell u-center-text") #tags and class are same for the all the teams except top position team
m
```

```python
M=[]
for i in m:
    M.append(i.text.replace('\n',''))
M
```

```python
len(M)
```

# MATCHES

```python
match=m[0:18:2]
match
```

```python
MATCHES=[]
for i in match:
    MATCHES.append(i.text.replace('\n',''))
MATCHES
```

```python
len(MATCHES)
```

# POINTS

```python
for i in MATCHES:
    try:
        M.remove(i)
    except ValueError:
        pass
M
```

```python
Points=M
Points
```

```python
len(Points)
```

# RATING

```python
r=soup.find_all('td', class_="table-body__cell u-text-right rating")
r
```

```python
rating=[]
for i in r:
    rating.append(i.text.replace('\n',''))
rating
```

```python
len(rating)
```

# 2nd Dataframe

```python
df_2=pd.DataFrame({})
df_2['TEAM']=T
df_2['MATCHES']=MATCHES
df_2['POINTS']=Points
df_2['RATING']=rating
df_2
```

# top_10 Womens_ODI_Teams_Rankings

```python

#combining the dtata frame 1 and & 2
Womens_ODI_Rankings=pd.concat([df_1,df_2],ignore_index=True)
Womens_ODI_Rankings
```
