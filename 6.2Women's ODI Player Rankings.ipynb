```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
soup.prettify
```

```python
#Name ODI Batting Rankings
n=soup.find('div', class_="rankings-block__banner--name-large") #ODI Batting Rankings
N=n.text.replace('\n','')
Na=N.split('\n')
Na
```

```python
#team
t=soup.find('div', class_="rankings-block__banner--nationality")
T=t.text.replace('\n','')
Te=T.split()
Te
```

```python
#team record
tr=soup.find('div', class_="rankings-block__career-best")
Tr=tr.text.replace('\n','')
Tr
```

```python
TR=Tr.split('\n')
TR
```

```python
#rating
r=soup.find('td', class_="u-text-left")
R=r.text.replace('\n','')
R
```

```python
rate=R.split()
rate
```

```python
import pandas as pd
df1=pd.DataFrame({})
df1['PLAYER']=Na
df1['TEAM']=Te
df1['TEAM_RECORD']=TR
df1['RATING']=rate
df1
```

```python
# creating data frame for remaining Women's ODI Batting Rankings
#name
N=soup.find_all('td', class_="table-body__cell rankings-table__name name")
N
```

```python
na=[]
for i in N:
    na.append(i.text.replace('\n',''))
na
```

```python
#team
te=soup.find_all('td',class_="table-body__cell nationality-logo rankings-table__team")
te
```

```python
tm=[]
for i in te:
    tm.append(i.text.replace('\n',''))
tm
```

```python
#team record
T=soup.find_all('td', class_="table-body__cell u-text-right u-hide-phablet")
T
```

```python
Tr=[]
for i in T:
    Tr.append(i.text.replace('\n',''))
Tr
```

```python
#rating
Ra=soup.find_all('td', class_="table-body__cell rating")
Ra
```

```python
rate=[]
for i in Ra:
    rate.append(i.text.replace('\n',''))
rate
```

# Women's ODI Batting Rankings

```python
#create 2nd Data Frame
df2=pd.DataFrame({})
df2['PLAYER']=na
df2['TEAM']=tm
df2['TEAM_RECORD']=Tr
df2['RATING']=rate
df2
```

```python
df3=pd.concat([df1,df2],ignore_index=True)
df3
```

# Top_10 Women's ODI Batting (players)

```python
df4=df3.iloc[0:10]
df4
```
