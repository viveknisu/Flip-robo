# Question_2

In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

```python
brand=[]
Name_of_product=[]
Rating=[]
No_of_Ratings=[]
Return_Exchange=[]
Expected_Delivery=[]
Availability=[]
O_details=[]
Price=[]
product_url=[]
```

```python
start=0
end=3
Url=[]
for i in range(start,end):    #scraping the details for the ist three page
    for i in driver.find_elements_by_xpath("//a[@class='a-size-base a-link-normal a-text-normal']"):
        Url.append(i.get_attribute('href'))
    
                                       #getting the link from the list for next page
        
    next_button=driver.find_elements_by_xpath("//li[@class='a-last']//a")
    try:
        driver.get(next_button[1].get_attribute('href'))
    except:
        driver.get(next_button[0].get_attribute('href'))
    
        
for i in Url:
    driver.get(i)
      
    try:      #brand details
        br=driver.find_element_by_xpath("//td[@class='a-span9']//span")
        brand.append(br.text)
    except NoSuchElementException:
        brand.append("-")
        
    
    try:   #product Name
        pd=driver.find_element_by_id("productTitle")
        Name_of_product.append(pd.text)
    except NoSuchElementException:
        Name_of_product.append("-")
        
        
    try: #product rating
        rate=driver.find_element_by_xpath("//span[@class='reviewCountTextLinkedHistogram noUnderline']")
        Rating.append((rate.get_attribute('textContent').replace('\n','')))
    except NoSuchElementException:
        Rating.append("-")
        
        
    try: #number of ratings
        nr=driver.find_element_by_id('acrCustomerReviewText')
        No_of_Ratings.append(nr.text)
    except NoSuchElementException:
        No_of_Ratings.append("-")
        
        
    try: #Return/Excahnge
        re=driver.find_element_by_id("RETURNS_POLICY")
        Return_Exchange.append(re.text)
    except NoSuchElementException:
        Return_Exchange.append("-")
        
        
    try: #Expected_delivery
        e_d=driver.find_element_by_id("ddmDeliveryMessage")
        Expected_Delivery.append(e_d.text)
    except NoSuchElementException:
         Expected_Delivery.append("-")
            
            
    try:   #Availability
        av=driver.find_element_by_id("availability")
        Availability.append(av.text)
    except NoSuchElementException:
        Availability.append("-")
        
    try:#other_details
        other=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']")
        O_details.append(other.text)
    except NoSuchElementException:
        O_details.append("-")
        
    try:#price
        p=driver.find_element_by_id("priceblock_ourprice")
        Price.append(p.text)   
    except NoSuchElementException:
        Price.append("-")
         
            
    try: #product_URL
        u=driver.find_element_by_xpath("//a[@class='a-link-normal sp_hqp_shared_adLink a-text-normal']")
        product_url.append(u.get_attribute("href"))
    except NoSuchElementException:
        product_url.append("-")
          
   
        
           
     
```

# Question_4

Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
#importing the require libraries
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
#connect to web driver
driver=webdriver.Chrome("chromedriver.exe")
driver
```

```python
url="https://www.flipkart.com/"
driver.get(url)
```

```python
#closinging the user loging id password option
driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']").click()
```

```python
#web element for the the searchbar
search=driver.find_element_by_class_name('_3704LK')
```

```python
#entering the searchvalue
search.send_keys('smart phones')
```

```python
#web element for the search buttion
driver.find_element_by_class_name('L0Z3Pu').click()
```

```python
smart_phone_name=[]
for i in driver.find_elements_by_class_name("_4rR01T"):
    smart_phone_name.append(i.text)
smart_phone_name    
```

```python
price=[]
for i in driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']"):
    price.append(i.text)
    
price
```

```python
#for the rest other information we have get inside the url of the page
```

```python
URL=[]
for i in driver.find_elements_by_xpath("//a[@class='_1fQZEK']"):
    URL.append(i.get_attribute('href'))
URL
```

```python
Brand=[]
color=[]
Ram=[]
camera=[]
display_size=[]
ROM=[]
Battery_capacity=[]
product_url=[]
processor=[]
```

```python
for i in URL:
    driver.get(i)
    
    try: #brand_name
        b=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[1]/div[1]/div/div[4]/a')
        Brand.append(b.text)
    except NoSuchElementException:
        Brand.append("-")
    
    try: #color
        b= driver.find_element_by_id('swatch-0-color')
        color.append(b.get_attribute('textContent'))
    except NoSuchElementException:
        color.append("-")
        
    try: #ram &ROm
        r=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[1]')
        Ram.append(r.text)
    except NoSuchElementException:
        Ram.append("_")
        
    try: #primary and 2ndry camera
        c=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[3]')
        camera.append(c.text)
    except NoSuchElementException:
        camera.append("-")
        
    try:#display size
        ds=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[2]')
        display_size.append(ds.text)  
    except NoSuchElementException:
        display_size.append("-")
        
    try: #ROM
        ro=driver.find_element_by_id('swatch-0-storage')
        ROM.append(ro.text)
    except NoSuchElementException:
        ROM.append("-")
        
    try: #battery_capacity
        bc= driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[4]')
        Battery_capacity.append(bc.text)
    except NoSuchElementException:
         Battery_capacity.append("-")
            
            
    try :#product_url
        pu=driver.find_element_by_id('swatch-0-storage')
        product_url.append(pu.get_attribute('href'))
    except NoSuchElementException:
        product_url.append("-")
        
    try: #processor
        pro=driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[5]')
        processor.append(pro.text)
    except NoSuchElementException:
        processor.append("-")
    
        
```

```python
df=pd.DataFrame({})
df['Brand Name']=Brand
df['Smart phone name']=smart_phone_name
df['Color']=color
df['RAM']=Ram
df['ROM']=ROM
df['primary,Secondary Camera']=camera
df['Prpcessor & Processor Cores']=processor
df['Display Size & Resolustion']=display_size
df['Battery_capacity']=Battery_capacity
df['Price']=price

df
```
