# ODI Team in mens cricket

#  teams
here we can clearly see the tags and classes are same for all the teams so we will apply find all for all 


```python
teams=soup.find_all('span',class_="u-hide-phablet")
teams
```

```python
ODI_mens_teams=[]
for i in teams:
    ODI_mens_teams.append(i.text.replace('\n',''))

ODI_mens_teams
    
```

```python
team=ODI_mens_teams[0:1]
team
```

```python
# Total_Matches
matches=soup.find('td',class_="rankings-block__banner--matches")
matches=soup.find('td',class_="rankings-block__banner--matches")
match=matches.text.replace('\n','')
match
```

```python
x=match.split()
x
```

```python
# points
points=soup.find('td',class_="rankings-block__banner--points")
point=points.text.replace('\n','')
point
```

```python
p=point.split()
p
```

```python
# rating
rating=soup.find('td',class_="rankings-block__banner--rating u-text-right")
rate=rating.text.replace('\n','')
rate
```

```python
r=rate.split()
r
```

```python
# creating the data frame for the top teams which is having diffrent tag and  classess in matches, rating and point columns
import pandas as pd
dt=pd.DataFrame({})
dt['TEAM']=team
dt['MATCHES']=x
dt['POINTS']=p
dt['RATING']=r
dt
```

# matches & Points.

```python
# matches and points
p2=soup.find_all('td', class_="table-body__cell u-center-text")
matches_Points=[]
for i in p2:
    matches_Points.append(i.text.replace('\n',''))
matches_Points
```

# Rating

```python
Rating=soup.find_all('td',class_="table-body__cell u-text-right rating")
Ratings=[]
for i in Rating:
    Ratings.append(i.text.replace('\n',''))
Ratings
```
