```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
soup.prettify
```

```python
#name
name=soup.find('div',"rankings-block__banner--name-large")
n=name.text.replace('\n','')
n
```

```python
N=n.split('\n')
N
```

```python
#team
t=soup.find('div', class_="rankings-block__banner--nationality")
T=t.text.replace('\n','')
T
```

```python
TE=T.split()
TE
```

```python
#Carrer best rating
cbr=soup.find('span', class_="rankings-block__career-best-text")
cb=cbr.text.replace('\n','')
cb
```

```python
Cb=cb.split('\n')
Cb
```

```python
#rating
rate=soup.find('td', class_="u-text-left")
Rate=rate.text.replace('\n','')
Rate
```

```python
R=Rate.split()
R
```

```python
import pandas as pd
df1=pd.DataFrame({})
df1['PLAYER']=N
df1['TEAM']=TE
df1['CAREER BEST RATING']=Cb
df1['RATING']=R
df1
```

#now we will scracp required data form all remaing having same tag and class and we will create 2nd data frame

```python
#name
n=soup.find_all('td', class_="table-body__cell rankings-table__name name")
n
```

```python
Names=[]
for i in n:
    Names.append(i.text.replace('\n',''))
Names
```

```python
#Team
t= soup.find_all('td', class_="table-body__cell nationality-logo rankings-table__team")
t
                            
```

```python
Teams=[]
for i in t:
    Teams.append(i.text.replace('\n',''))
Teams
```

```python
#Rating
rate=soup.find_all('td', class_="table-body__cell rating")
rate
```

```python
Rating=[]
for i in rate:
    Rating.append(i.text.replace('\n',''))
Rating
```

```python
#CAREER BEST RATING
crb=soup.find_all('td', class_="table-body__cell u-text-right u-hide-phablet")
crb
```

```python
CAREER_BEST_RATING=[]
for i in crb:
    CAREER_BEST_RATING.append(i.text.replace('\n',''))
CAREER_BEST_RATING
```

# Creating the 2nd data frame

```python
Mens_ODI_Bowling_Rankings=pd.DataFrame({})
Mens_ODI_Bowling_Rankings['PLAYER']=Names
Mens_ODI_Bowling_Rankings['TEAM']=Teams
Mens_ODI_Bowling_Rankings['CAREER BEST RATING']=CAREER_BEST_RATING
Mens_ODI_Bowling_Rankings['RATING']=Rating
Mens_ODI_Bowling_Rankings
```

```python
# combining the both data frame
dc=pd.concat([df1,Mens_ODI_Bowling_Rankings],ignore_index=True)
dc
```
