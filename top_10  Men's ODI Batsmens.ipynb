```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
soup.prettify
```

# as we can see the tags and classes of the top position player is diffrent from rest other so we will scrap all the data seprately 

```python
#player_name
n1=soup.find('div',class_="rankings-block__banner--name-large")
N=n1.text.replace('\n','')
N
```

```python
Na=N.split('\n')
Na
```

```python
#Team_name
t=soup.find('div',class_="rankings-block__banner--nationality")
t1=t.text.replace('\n','')
t1
```

```python
TN=t1.split()
TN
```

```python
#team_record
tr=soup.find('div',class_="rankings-block__career-best")
TR=tr.text.replace('\n','')
TR
```

```python
Tr=TR.split('\n')
Tr
```

```python
#Rating
rate=soup.find('td',class_="u-text-left")
Rate=rate.text.replace('\n','')
Rate
```

```python
R=Rate.split()
R
```

# creating data frame for the ist ranking batsman
which is having diffrent tags and class

```python
import pandas as pd
First=pd.DataFrame({})
First['PLAYER']=Na
First['TEAM']=TN
First['CAREER BEST RATING']=Tr
First['RATING']=R
First
```

# Creating 2nd DataFrame 
Creating Data Frame for all the teams which having same tag and class

```python
n2=soup.find_all('td',class_="table-body__cell rankings-table__name name")
n2
```

```python
name_list=[]
for i in n2:
    name_list.append(i.text.replace('\n',''))
name_list
```

```python
teamlist=soup.find_all('span',class_=("table-body__logo-text"))
teamlist
```

```python
TEAM=[]
for i in teamlist:
    TEAM.append(i.text.replace('\n',''))
TEAM
```

```python
CarrerRating=soup.find_all('td',class_=("table-body__cell u-text-right u-hide-phablet"))
CarrerRating
```

```python
CAREER_BEST_RATING=[]
for i in CarrerRating:
    CAREER_BEST_RATING.append(i.text.replace('\n',''))
CAREER_BEST_RATING
    
```

```python
rate=soup.find_all('td',class_="table-body__cell rating")
rate
```

```python
RATING=[]
for i in rate:
    RATING.append(i.text.replace('\n',''))
RATING
```

```python
Top_10_ODI_Batsmen=pd.DataFrame({})
Top_10_ODI_Batsmen['PLAYER']=name_list
Top_10_ODI_Batsmen['TEAM']=TEAM
Top_10_ODI_Batsmen['CAREER BEST RATING']=CAREER_BEST_RATING
Top_10_ODI_Batsmen['RATING']=RATING
Top_10_ODI_Batsmen
```

```python
#combining dataframe 1 and 2
frame=[First,Top_10_ODI_Batsmen]
Top_10_ODI=pd.concat([First,Top_10_ODI_Batsmen],ignore_index=True)
Top_10_ODI
```
