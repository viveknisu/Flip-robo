# importing required libararies

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/reviews/26173-joan-silber-secrets-happiness-fiction#.YJKNbsDivIU")
page
```

```python
page.content
```

```python
soup=(BeautifulSoup(page.content))
soup
```

```python
print(soup.prettify)
```

# 1st book name,Author name,genre,review,

```python
name=soup.find('h4',class_='italic')
name
```

```python
name.text.replace('\n','')
```

#author_name

```python
author=soup.find('p',class_="sans bold")
author
```

```python
author.text.replace('\n','')
```

# genre

```python
genre=soup.find('p',class_="genre-links")
genre
```

```python
genre.text.replace('\n','')
```

# review.

```python
review=soup.find('div',class_="span7")
review
```

```python
review.text.replace('\n',' ')
```
