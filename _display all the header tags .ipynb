# Task Details:
            Write a python program to display all the header tags  from web site‘en.wikipedia.org/wiki/Main_Page’.

# Importing the Required Libraries

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify)
```

```python
title=soup.find_all(['hi','h2','h3','h4','h5','h6'])
title
```

```python
print('all header tags are :',*title,sep='\n\n')
```
