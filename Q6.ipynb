```python
import pandas as pd
import selenium
from selenium import webdriver
```

```python
#connect to web driver
driver=webdriver.Chrome("chromedriver.exe")
driver
```

```python
url="https://www.flipkart.com/"
driver.get(url)
```

```python
#seach bar
x=driver.find_element_by_xpath("//div[@class='_3OO5Xc']//input")
x
```

```python
x.send_keys('sunglasses')
```

```python
#seachbutton
driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
```

```python
brand=[]
for b in driver.find_elements_by_xpath("//div[@class='_2WkVRV']"):
    brand.append(b.text)
brand    
```

```python
prod_Desc=[]
for d in driver.find_elements_by_xpath("//div[@class='_2B099V']//a"):
    prod_Desc.append(d.text)
prod_Desc
```

```python
prod_Desc[0:79:2]
```

```python
len(prod_Desc[0:79:2])
```

```python
price=[]
for p in driver.find_elements_by_xpath("//div[@class='_30jeq3']"):
    price.append(p.text)
price    
```

```python
dis_prcnt=[]
for p in driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']"):
    dis_prcnt.append(p.text)
dis_prcnt
```

```python
df=pd.DataFrame({})
df[' Brand']=brand  
df['Product Description']=prod_Desc[0:79:2]
df[' Price']=price 
df['Discount %']=dis_prcnt
df
```
